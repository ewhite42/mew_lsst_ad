```python
import scienceplots
import matplotlib.pyplot as plt

import matplotlib.colors as colors
from matplotlib.patches import Patch
import seaborn as sns

import pandas as pd
import numpy as np

from sklearn.neighbors import NearestNeighbors
from sklearn.decomposition import PCA

plt.style.use("science")
```

```python
## Load in the normalised test data, the encoded data, and the decoded data -MEW
X = np.load("../data/normalised_objs.npy")
encoded = np.load("../data/encoded_objs.npy")
decoded = np.load("../data/decoded_objs.npy")

## calculate the reconstruction losses then delete the decoded and X objects -MEW
recon_losses = np.sum((X - decoded)**2, axis=1)
del decoded, X
```

```python
data = pd.read_csv("../../dp03/dp03_objects.csv")
# Clean
## calculate the semi-major axis from the q and e columns, and add that to the pandas dataframe -MEW
data["a"] = data["q"]/(1 - data["e"])

## remove rows with NA values and remove the specified columns -MEW
clean_objs = data.dropna()
clean_objs = clean_objs.drop(columns=["ssObjectId", "fullDesignation", "q2", "e2", "Unnamed: 0"])

## calculate and add new columns to the cleaned dataframe -MEW
clean_objs["i-z"] = clean_objs["i_H"] - clean_objs["z_H"]
clean_objs["g-r"] = clean_objs["g_H"] - clean_objs["r_H"]
clean_objs["r-i"] = clean_objs["r_H"] - clean_objs["i_H"]

clean_objs["a*"] = .89*(clean_objs["g-r"]) + .45*(clean_objs["r-i"]) -.57

clean_objs["a"] = clean_objs["q"]/(1-clean_objs["e"])

clean_objs["Q"] = clean_objs["a"]*(1 + clean_objs["e"])
clean_objs["T_jup"] = (5.204/clean_objs["a"]) + 2*np.cos((clean_objs["incl"]/180)*np.pi)*np.sqrt((clean_objs["a"]/5.204)*(1 - clean_objs["e"]**2))

## calculate and print the fraction of data that's left after cleaning it -MEW
print((len(data) - len(clean_objs))/len(data)) # Dropped
```

```python
## fit the nearest neighbors estimator to the encoded data -MEW
nn = NearestNeighbors(n_neighbors=20).fit(encoded)
```

```python
anoms = []

## for the ten most anomalous data points (based on largest reconstruction
## losses), append the index of the 20 corresponding nearest neighbors in 
## the k nearest neighbors array (nn) fit above -MEW
for obj in np.argsort(recon_losses)[-10::][::-1]:
    kn_idx = nn.kneighbors([encoded[obj]])
    
    anoms.append(
        kn_idx[1][0]
    )
    
#iso_index = np.argsort(recon_losses)[-20::][::-1][5]

#kn_idx = nn.kneighbors([encoded[iso_index]])
```

```python
## print full designation of 10 most anomalous objects
print(data.loc[clean_objs.iloc[np.argsort(recon_losses)[-10::][::-1], :].index, :]["fullDesignation"])
```

```python
## define dataframe containing the 6-dimensional encoded data -MEW
df = pd.DataFrame(
    encoded,
    columns = [f"AE {i}" for i in range(1,7)]
)

## initialize "cluster" number in dataframe to zero -MEW
df["cluster"] = 0

## update the "cluster" values in each row so that the first cluster 
## is labelled 1, the second cluster is labelled 2, and so on -MEW
for i, clst_idx in enumerate(anoms):
    df.loc[clst_idx, "cluster"] = i + 1
```

```python
values = np.arange(11)

# Choose a discrete color palette
colors = plt.cm.tab20(values)[1:]

## use seaborn pairplot to plot the anomaly clusters against each component of the encoded data -MEW
pp = sns.pairplot(
    df[[f"AE {i}" for i in range(1,7)]],
    plot_kws={
        "s":0.01, 
        "alpha":0.15,
        },
    corner=True,
    diag_kind=None
)

## set the legend -MEW 
legend_elements = [Patch(facecolor=color,
                         label=f'{i + 1}') for i, color in enumerate(colors)]
legend_elements = [Patch(facecolor="blue", label=f'Normal')] + legend_elements
pp.fig.legend(handles=legend_elements, bbox_to_anchor=(0.8, .85), fontsize=20)

## plot the data in each of the pairwise plots and set the axis names -MEW
for ax in pp.axes.flat:
    for _, (clst_idx, color) in enumerate(zip(anoms, colors)):
        for i, axs in enumerate(pp.axes):
            for j, ax in enumerate(axs):
                if ax:
                    if i != j:
                        ax.scatter(
                            df.iloc[clst_idx, j],
                            df.iloc[clst_idx, i],
                            s=2,
                            c=color
                        )
                        ax.set_xlabel(f"AE {j + 1}", fontsize=14)
                        ax.set_ylabel(f"AE {i + 1}", fontsize=14)
                    elif i == j:
                        ax.set_visible(False)
                    else:
                        pass
plt.show()
```

```python
## transform the encoded data from 6 dimensions to 3 using a PCA in order to plot it -MEW
pca = PCA(n_components = 3).fit_transform(
    encoded
)
```

```python
ax = plt.figure(figsize=(12,12)).add_subplot(projection='3d')

## transform the 20 cluster 6 data points to 3-d space using pca -MEW
clst6_nn = pca[df[df["cluster"] == 6].index, :]

## save all 20 datapoints of cluster 6 -MEW
clst6 = df[df["cluster"] == 6]

## add columns to the cluster 6 dataframe such as: reconstruction loss (rl), name, 
## eccentricity (e), and sort by reconstruction loss -MEW
clst6["rl"] = recon_losses[df[df["cluster"] == 6].index]
clst6["name"] = data.loc[clean_objs.iloc[df[df["cluster"] == 6].index, :].index, :]["fullDesignation"].values
clst6["e"] = clean_objs.iloc[df[df["cluster"] == 6].index, :]["e"].values
clst6 = clst6.sort_values(by="rl")

## plot the least anomalous of the 20 cluster 6 objects first -MEW
ax.scatter3D(
        clst6_nn[0, 0], 
        clst6_nn[0, 1],
        zs=clst6_nn[0, 2],
        c="red",
        edgecolor="black",
        s=40
    )

## iterate through the next of the 20 cluster objects -MEW
for i, name, e in zip(range(1,20), clst6["name"], clst6["e"]):
    ## plot a dashed line leading from the least anomalous of the 20 objects
    ## to the current object's marker -MEW
    ax.plot(
        [clst6_nn[0, 0], clst6_nn[i, 0]], 
        [clst6_nn[0, 1], clst6_nn[i, 1]],
        zs=[clst6_nn[0, 2], clst6_nn[i, 2]],
        c="black",
        alpha=.2,
        ls="--"
        )
    ## plot the current object's location in the 3-d space
    ax.scatter3D(
        clst6_nn[i, 0], 
        clst6_nn[i, 1],
        zs=clst6_nn[i, 2],
        c="r",
        edgecolor="black",
    )
    
## plot the rest of the datapoints in 3-d space
ax.scatter(pca[:, 0], pca[:, 1], zs=pca[:, 2],alpha=.05,
    s=.001,
    marker="o",
    c="blue",
    )

ax.view_init(elev=20., azim=-60, roll=0)



ax.grid(True)
ax.set_facecolor("white")

plt.tight_layout()
plt.tight_layout()

ax.grid(False)
ax.set_xlabel('PCA 1', fontsize=20, labelpad=5)
ax.set_ylabel('PCA 2', fontsize=20, labelpad=5)
ax.set_zlabel('PCA 3', fontsize=20, labelpad=-30)

#plt.savefig("ISO_network_pca.png", dpi=300)
plt.show()
```
